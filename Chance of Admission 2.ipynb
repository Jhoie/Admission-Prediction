```python
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Admission_Predict.csv")
df
```

```python
df.drop("Serial No.", axis=1, inplace=True)
```

```python
df.rename(columns={'Chance of Admit ':'ChanceofAdmit', 'GRE Score':'GREScore', 'TOEFL Score':'TOEFLScore',\
                  'University Rating':'UniversityRating', 'LOR ':'LOR'}, inplace=True)
```

```python
GRE_mean = df.GREScore.mean()
df['GRE_centered'] = df.GREScore - GRE_mean

TOEFL_mean = df.TOEFLScore.mean()
df['TOEFL_centered'] = df.TOEFLScore - TOEFL_mean

UniversityRating_mean = df.UniversityRating.mean()
df['UniversityRating_centered'] = df.UniversityRating - UniversityRating_mean

SOP_mean = df.SOP.mean()
df['SOP_centered'] = df.SOP - SOP_mean

LOR_mean = df.LOR.mean()
df['LOR_centered'] = df.LOR - LOR_mean

```

```python
multiLinearModel = smf.ols(formula='ChanceofAdmit ~ GRE_centered + TOEFL_centered + UniversityRating_centered + LOR_centered \
                           + Research + CGPA', data=df)
multiLinearModResult = multiLinearModel.fit()
print(multiLinearModResult.summary())
```
