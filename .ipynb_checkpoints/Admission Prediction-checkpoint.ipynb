# VISUALIZATIONS

### Checking the distribution for the target variable(Chance of Admit)


```python
sns.displot(df["Chance of Admit "], kde = True)
```

```python
# The distribution is skewed to the left meaning the mean and median will be less than the mode.
```

```python
Admit_mode = df["Chance of Admit "].mode()
Admit_mode
```

```python
Admit_median = df["Chance of Admit "].median()
Admit_median
```

```python
Admit_mean = df["Chance of Admit "].mean()
Admit_mean
```

```python
# BUT THE DEFINITION OF LEFT SKEWNESS IS DIFFERENT FROM WHAT I AM GETTING, COULD TGERE BE SOMETHING WRONG?
# The mode is supposed to be greater than the mean and median but the opposite is the case
```

# VISUALIZING THE GRE Score feature

```python
sns.displot(data=df, x="GRE Score") # The most frequently occuring score is 325 with over 60 students getting that score
```

```python
sns.displot(data=df, x="GRE Score", kind = "kde")
```

```python
sns.lineplot(data=df, x="GRE Score", y="Chance of Admit ") # the line plot below shows that a higher GRE score gives a higher chance of admit
```

```python
sns.lineplot(data=df, x="GRE Score", y="Chance of Admit ", hue = "Research") # majority of the time, a reasearch also increases your chances of getting admitted
```

```python
sns.lineplot(data=df, x="GRE Score", y="Chance of Admit ", hue = "University Rating") # if your university rating is 5, you have at least 0.6 of gettting admit
```

```python
sns.scatterplot(data=df, x="GRE Score", y="Chance of Admit ", hue="Research", style = "Research") # The scatter plot also shows that a Research increases the chances of a admit
```
